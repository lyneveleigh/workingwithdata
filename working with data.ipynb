```python
#merge 

df1 = DataFrame({'key':['x','x','y','y','z','y'],'data':np.arange(6)})
df1
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

```python
pd.merge(df1,df2,on='key',how='left') #pass them using the left data
```

```python
df_left = DataFrame({'key1':['NYC','BS','LA'],
                    'key2':['one','two','three'],
                    'left_data':[10,20,30]})

df_right = DataFrame({'key1':['NYC','BS','LA','LA'],
                    'key2':['one','two','three','two'],
                    'left_data':[30,50,60,70]})
```

```python
df_left
```

```python
df_right
```

```python
pd.merge(df_left,df_right,on=['key1','key2'],how='outer')
```

```python
pd.merge(df_left,df_right,on=['key1'])
```

```python
pd.merge(df_left,df_right,on=['key1'],suffixes=('_lefty','_righty'))
```
